```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
# Loading data
file_path = "../Part 1/master_df.csv"
df_genre = pd.read_csv(file_path)
df_genre.head(10)
```

```python
df_genre.dtypes
```

```python
x = df_genre[['popularity', 'duration', 'mode', 'speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
```

```python
# Initializing model with K = 4 (since we already know there are four genres)
model = KMeans(n_clusters=4, random_state=5)
model
```

```python
# Fitting model
model.fit(x)
```

```python
# Get the predictions
predictions = model.predict(x)
print(predictions)
```
