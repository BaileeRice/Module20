```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#adding connection string
from sqlalchemy import create_engine
from config import config
#adding connection string
engine = create_engine(f'postgresql://postgres:{config}@localhost:5432/Module20')
```

```python
# Loading data
df_genre = pd.read_sql('SELECT * FROM master', engine)
df_genre.head(10)
```

```python
df_genre.dtypes
```

# Building the KMeans Model

```python
x = df_genre[['popularity', 'duration', 'mode', 'speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
```

```python
# Initializing model with K = 4 (since we already know there are four genres)
model = KMeans(n_clusters=4, random_state=5)
model
```

```python
# Fitting model
model.fit(x)
```

```python
# Get the predictions
predictions = model.predict(x)
print(predictions)
```

```python
# Add a new class column to the df
df_genre["class"] = model.labels_
df_genre.head()
```

```python
df_genre['track_id'].value_counts()
```

```python
# Create a scatterplot of df_genre
df_genre.hvplot.scatter(x="track_id", y="genre", by="class")
```

# PCA

```python
#standardize data with standardscaler
genre_scaled = StandardScaler().fit_transform(x)
print(genre_scaled[0:5])
```

```python
# Initialize PCA model
pca = PCA(n_components=3)
```

```python
# Get principal components for the genre data.
genre_pca = pca.fit_transform(genre_scaled)
```

```python
df_genre_pca = pd.DataFrame(
    data=genre_pca, columns=['principal component 1', 'principal component 2', 'principal component 3']
)
df_genre_pca.head()
```

```python
# Add the predicted class columns
df_genre_pca["class"] = df_genre['genre']
df_genre_pca.head()
```

```python
# Use explained_variance_ratio to learn how much information can be attributed to each principal component:
pca.explained_variance_ratio_
```

```python
# Plotting the clusters with three features
fig = px.scatter_3d(df_genre_pca, x="principal component 1", y="principal component 2", z="principal component 3", color='class', width=800)
fig.update_layout(legend=dict(x=0,y=1))
fig.show()
```

```python
# Initializing model with K = 4 (since we already know there are three genres)
model = KMeans(n_clusters=4, random_state=5)
model
```

```python
# Fitting model
model.fit(df_genre_pca[["principal component 1", "principal component 2", "principal component 3"]])
```

```python
# Get the predictions
predictions = model.predict(df_genre_pca[["principal component 1", "principal component 2", "principal component 3"]])
print(predictions)
```

```python
# Add the predicted class columns
df_genre_pca["class_2"] = model.labels_
df_genre_pca.head()
```

```python
fig = px.scatter_3d(df_genre_pca, x="principal component 1", y="principal component 2", z="principal component 3", color='class_2', width=800)
fig.update_layout(legend=dict(x=0,y=1))
fig.show()
```
